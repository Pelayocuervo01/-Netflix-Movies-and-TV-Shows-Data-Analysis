**Data Acquisition and Library Imports**

---



*   Importing Libraries: First we import Python libraries such as Pandas, NumPy, Seaborn, and Matplotlib, which will be utilized for data manipulation, analysis, and visualization throughout the project.
*   Loading the Dataset: This step involves loading the Netflix movies dataset from a CSV file named "netflix_movies.csv", located in the project's file directory, by utilizing the Pandas library. The CSV file content is then parsed and stored within a Pandas DataFrame called "df", to be readily available for use in the following project sections.








```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
database = pd.read_csv("netflix_movies.csv")
df = pd.DataFrame(database)
df
```

**Extracting Date and creating new columns**

---



*   Extracting Month: The original date added string is manipulated with Pandas and string replacement methods, to extract the month from the 'date_added' column.
*   Extracting Day: Similar string manipulation using regular expressions allows the extraction of the day of the month from the 'date_added' column.
*   Extracting Year: With Pandas string extraction techniques, the year is also obtained from the same 'date_added' column.

These extracted components are then stored as dedicated columns namely 'month_added', 'day_added', and 'year_added' within the DataFrame, providing a more structured representation of the movie addition date for improved analysis.



```python
df.info()
```

```python
df["month_added"] = df["date_added"].str.replace(r"[0-9,]","", regex=True)
df.head(2)
```

```python
df["day_added"] = df["date_added"].str.replace(r"[a-zA-Z,]","",regex=True).str.extract(r'(\d{2})')
df["year_added"] = df["date_added"].str.replace(r"[a-zA-Z,]","",regex=True).str.extract(r'(\d{4})')
df.head(2)
```

**Handling Missing Date Added Values**


---


*   Identifying and Handling NaN Values: Instances where the 'date_added' column contains missing values (represented as NaN) are identified using Pandas isna() and where() methods.
*   Filling NaN Values with 'release_year': To address the identified NaNs in the 'date_added' column, it is selectively replaced by the movie's release year (contained in the 'release_year' column) providing a close alternative in the absence of original addition date information.




```python
df[df["month_added"].isna()].head(2)
```

```python
df["date_added"] = np.where(df["date_added"].isna(), df["release_year"], df["date_added"])
df.head(2)
```

```python
df[df["duration"].isna()]
```

**Inconsistent Data in duration Column**


---

During data exploration, it was observed that in some instances where the 'duration' column had missing values (NaN), the actual duration value was mistakenly placed in the 'rating' column.  To address this inconsistency, these values were transferred from the 'rating' column to the 'duration' column, ensuring data accuracy.


```python
df["duration"] = np.where(df["duration"].isna(), df["rating"], df["duration"])
```

```python
df[df["country"].isna()].head()
```

**Filling Missing Country and Director Info**



---

For movies with missing country or director info (those pesky NaNs!), we filled them with the most common country and director in the dataset. This way, we have something instead of nothing!


```python
df["country"].fillna(df["country"].value_counts()[0],inplace=True)
df["director"].fillna(df["director"].value_counts()[0],inplace=True)
```

**Breaking Down Show Categories and Visualizing Trends**


---

We split the 'listed_in' column to see all the different show categories and find out which ones are the most popular. Then, we created some cool charts to help us visualize this information:

1.  **Show Category Distribution:** This bar chart shows how many shows are in each category, helping us quickly see which categories are most common on Netflix.
2.  **Shows Added Over Time:** This bar chart displays how many shows were added each year, giving us a sense of how Netflix's content has grown.
3.  **Movies vs. Shows Over Time:** This stacked bar chart breaks down the content added each year into movies and TV shows, allowing us to compare their growth trends.


```python
show_category = []
for i in df["listed_in"].str.split(","):
  show_category.append(i[0])
df["show_category"] = show_category
```

```python
########################## https://medium.com/@BrendanArtley/matplotlib-color-gradients-21374910584b #######################################
color1 = "#8A5AC2"
color2 = "#3575D5"
def hex_to_RGB(hex_str):
    """ #FFFFFF -> [255,255,255]"""
    #Pass 16 to the integer function for change of base
    return [int(hex_str[i:i+2], 16) for i in range(1,6,2)]

def get_color_gradient(c1, c2, n):
    """
    Given two hex colors, returns a color gradient
    with n colors.
    """
    assert n > 1
    c1_rgb = np.array(hex_to_RGB(c1))/255
    c2_rgb = np.array(hex_to_RGB(c2))/255
    mix_pcts = [x/(n-1) for x in range(n)]
    rgb_colors = [((1-mix)*c1_rgb + (mix*c2_rgb)) for mix in mix_pcts]
    return ["#" + "".join([format(int(round(val*255)), "02x") for val in item]) for item in rgb_colors]
##############################################################################################################################################


show_category_counts = df["show_category"].value_counts()

plt.figure(figsize=(12,7))
plt.bar(show_category_counts.index, show_category_counts.values,color=get_color_gradient(color1, color2, len(show_category_counts)))
plt.xlabel("Shows Categories")
plt.ylabel("Number of Shows")
plt.title("Distribution of Show Categories")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
year_list = df.groupby(["year_added"])[["show_category"]].count()
year_list = year_list.sort_values(by=["year_added"], ascending=False)
```

```python
plt.figure(figsize=(12, 7))
plt.bar(year_list.index, year_list["show_category"].values, color=get_color_gradient(color1, color2, len(year_list)),align='edge')
plt.xlabel("Year Added")
plt.ylabel("Number of Shows Added")
plt.title("Distribution of Shows Added Over the Years")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
year_type = df.groupby(["year_added","type"])[["type"]].count()
year_type
```

```python
year_type.columns = ["count"]
year_type = year_type.reset_index()


year_type_pivot = year_type.pivot(index="year_added", columns="type", values="count")

year_type_pivot.plot(kind="bar", stacked=True, figsize=(12, 7), color=[color1, color2])
plt.xlabel("Year Added")
plt.ylabel("Number of Shows and Movies")
plt.title("Distribution of Shows and Movies Added Over the Years")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Type")
plt.tight_layout()
plt.show()
```
